# Aqui começa a parte pro arquivo todo

```python
reader = PdfReader('CLIENTES DOCTORCIA.pdf')
```

```python
lista_estados =[]
for page in range(len([x for x in reader.pages])):
    print(page)
    text = reader.pages[page].extract_text()
    texto = text.splitlines()

    
    for x in texto:
        try:
            lista_estados.append(re.findall('\d+/\d+/\d+[ ]*(.*-\w+)', x)[0].strip().replace('  ', ''))
        except:
            continue


# Limpa os tracos e espacos no comeco e fim das palavras    
lista_estados_final = [(re.findall('.*-[A-Z][A-Z]', x)[0]).strip('- ') for x in lista_estados]

```

```python
lista_estados_final
```

```python
with open('resultado.txt', 'w+') as f:
    for linha in lista_estados_final:
        # Retira os que só tem nome de estado, sem cidade
        if len(linha) > 2:
            f.write(f'{linha}\n')
```
